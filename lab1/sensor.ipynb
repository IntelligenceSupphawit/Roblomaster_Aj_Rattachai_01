```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sensor_data.csv")
df.set_index('timestamp', inplace=True)  
df
```

```python
graph_speed = df['speed']
graph_speed = graph_speed.dropna()

plt.title('Speed Over Time')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.plot(graph_speed)
plt.show()
```

```python
graph_gyro = df[['gyro_x', 'gyro_y', 'gyro_z']]
graph_gyro = graph_gyro.dropna()

plt.title('Gyroscope')
plt.xlabel('Time')
plt.ylabel('Gyro') 
plt.plot(graph_gyro)
plt.show()
```
