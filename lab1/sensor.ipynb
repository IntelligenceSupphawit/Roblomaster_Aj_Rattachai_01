```python
import pandas as pd

columns = [
    "timestamp", "speed", "angle",
    "yaw", "pitch", "roll",
    "acc_x", "acc_y", "acc_z",
    "gyro_x", "gyro_y", "gyro_z",
    "x", "y", "z",
    "static_flag", "impact_x", "impact_y", "impact_z",
    "up_hill", "down_hill", "on_slope"
]

# อ่านไฟล์แบบบังคับให้ค่าทั้งหมดเป็น string ก่อน
df = pd.read_csv("sensor_data.csv", header=None, names=columns, dtype=str)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors='coerce')
# แปลงคอลัมน์ตัวเลขให้กลายเป็น float ทีหลัง
float_cols = [
    "speed", "yaw", "pitch", "roll",
    "acc_x", "acc_y", "acc_z",
    "gyro_x", "gyro_y", "gyro_z",
    "x", "y", "z"
]

for col in float_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # ถ้าแปลงไม่ได้จะกลายเป็น NaN

# ทดสอบพล็อตค่าความเร่ง
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
# plt.plot(df.index, df["static_flag"], label="1")
# plt.plot(df.index, df["up_hill"], label="2")
# plt.plot(df.index, df["down_hill"], label="3")
# plt.plot(df.index, df["on_slope"], label="4")
plt.plot(df.index, df["impact_x"], label="6")
plt.plot(df.index, df["impact_y"], label="7")
plt.plot(df.index, df["impact_z"], label="8")
plt.title("--")
plt.xlabel("เวลา")
plt.ylabel("--")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # หมุนแกนเวลาให้ดูง่าย
plt.tight_layout()
plt.show()
```
